# 1.Data Loading & Initial Inspection
Goal: Load the dataset, inspect basic structure, and check for missing/duplicate values.

Explanation:

1. Understand the number of rows and columns
2. Get summary statistics for numerical features
3. Identify missing values and duplicates for proper cleaning


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/pakistans-largest-e-commerce-dataset/Pakistan Largest Ecommerce Dataset.csv")
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
import numpy as np
df = df.replace([np.inf, -np.inf], np.nan).fillna(0)
df.describe()
```

```python
import numpy as np
df = df.replace([np.inf, -np.inf], np.nan).fillna(0)
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(f"Duplicate Rows: {df.duplicated().sum()}")
```

# 2. Data Cleaning

Goal: Remove unnecessary columns, fix data types, and handle missing or invalid values.

Explanation:

1. Drop unnamed columns generated during CSV export
2. Convert created_at and Customer Since to datetime
3. Remove rows missing critical info like price, qty_ordered, or grand_total
4. Filter out zero or negative prices to ensure data quality

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
df['Customer Since'] = pd.to_datetime(df['Customer Since'], errors='coerce')
```

```python
df.dropna(subset=['item_id', 'status', 'price', 'qty_ordered', 'grand_total'], inplace=True)
```

```python
df = df[df['price'] > 0]
```

# 3. Feature Engineering

Goal: Create additional features to enable deeper analysis.

Explanation:

1. Calculate total order price
2. Flag discounted orders
3. Extract year, month, weekday from order dates
4. Compute customer tenure (days since first purchase)

```python
df['TotalPrice'] = df['price'] * df['qty_ordered']
```

```python
df['IsDiscounted'] = (df['discount_amount'] > 0).astype(int)
```

```python
df['OrderYear'] = df['created_at'].dt.year
df['OrderMonth'] = df['created_at'].dt.month
df['OrderWeekday'] = df['created_at'].dt.day_name()
```

```python
df['CustomerTenureDays'] = (df['created_at'] - df['Customer Since']).dt.days
```

# 4. Univariate Analysis

Goal: Understand distributions of individual features and identify outliers.

Explanation:


1. Visualize TotalPrice distribution
2. Count orders by status
3. Examine popularity of payment methods


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['price'], kde=True)
plt.title("Total Price Distribution")
plt.show()
```

```python
sns.countplot(y='status', data=df, order=df['status'].value_counts().index)
plt.title("Order Status Count")
plt.show()
```

```python
sns.countplot(y='payment_method', data=df, order=df['payment_method'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Payment Method Usage")
plt.show()
```

# 5. Bivariate Analysis

Goal: Explore relationships between two variables.

Explanation:

1. How payment method affects order status
2. How product category relates to total order value

```python
sns.countplot(y='payment_method', hue='status', data=df)
plt.xticks(rotation=45)
plt.title("Payment Method vs Order Status")
plt.show()
```

```python
sns.boxplot(x='category_name_1', y='TotalPrice', data=df)
plt.xticks(rotation=90)
plt.title("Category vs Total Price")
plt.show()
```

# 6. Multivariate Analysis

Goal: Identify correlations and patterns among multiple variables.

Explanation:

1. Numerical correlations to detect strong relationships
2. Pivot table to see monthly sales by category

```python
num_cols = ['price', 'qty_ordered', 'grand_total', 'discount_amount', 'TotalPrice', 'CustomerTenureDays']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Numerical Features Correlation")
plt.show()
```

```python
monthly_sales = pd.pivot_table(df, index='OrderMonth', columns='category_name_1', values='TotalPrice', aggfunc='sum')
monthly_sales.plot(kind='line', figsize=(20,6))
plt.title("Monthly Sales by Category")
plt.show()
```

# 7. Seasonal & Trend Analysis

Goal: Explore temporal patterns and seasonality.

Explanation:

1. Trend of sales over time
2. Weekday-wise sales distribution

```python
time_sales = df.groupby('created_at')['TotalPrice'].sum().reset_index()
plt.figure(figsize=(14,6))
sns.lineplot(x='created_at', y='TotalPrice', data=time_sales)
plt.title("Sales Trend Over Time")
plt.show()
```

```python
weekday_sales = df.groupby('OrderWeekday')['TotalPrice'].sum().reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
weekday_sales.plot(kind='bar')
plt.title("Sales by Weekday")
plt.show()
```

# 8. Cancellation & Payment Analysis

Goal: Examine cancellations and payment success rates.

Explanation:

1. Cancellation rate by payment method
2. Impact of discounts on order value

```python
cancel_rates = df.groupby('payment_method')['status'].value_counts(normalize=True).unstack().fillna(0)
print(cancel_rates)
```

```python
sns.boxplot(x='IsDiscounted', y='TotalPrice', data=df)
plt.title("Discount Impact on Total Price")
plt.show()
```

# 9. Customer Insights
Goal: Identify high-value and loyal customers.

Explanation:

1. Top customers based on spending
2. Calculate repeat purchase rate

```python
top_customers = df.groupby('Customer ID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
repeat_customers = df['Customer ID'].value_counts()
repeat_rate = (repeat_customers[repeat_customers > 1].count() / df['Customer ID'].nunique()) * 100
print(f"Repeat Purchase Rate: {repeat_rate:.2f}%")
```
